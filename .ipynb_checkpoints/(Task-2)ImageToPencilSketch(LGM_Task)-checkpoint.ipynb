```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
# Reading the image and plotting the original image 

img = cv2.imread('parliament.jpeg')
color_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(color_img)
plt.title('parliament building')
plt.axis('off')
plt.show()
```

```python
# printing the grey scale image of the above image 

grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img , cmap = 'gray')
plt.axis('off')
plt.title('gray image')
plt.show()
```

```python
# printing the inverted image 

inverted_image = 255 - grey_img 
plt.imshow(inverted_image , cmap = 'gray')
plt.axis('off')
plt.title('inverted image')
plt.show()
```

```python
# printing the blurred image 

blured_img = cv2.GaussianBlur(inverted_image, (21,21),0)
plt.imshow(blured_img , cmap = 'gray')
plt.axis('off')
plt.title('blurred image')
plt.show()
```

```python
# inverting the blurred image 

inverted_blurred_img = 255 - blured_img 
plt.imshow(inverted_blurred_img , cmap = 'gray')
plt.axis('off')
plt.title('inverted blurred image')
plt.show()
```

```python
# creating the pencil sketch 

pencil_sketch = cv2.divide(grey_img,inverted_blurred_img,scale = 256.0)
plt.imshow(pencil_sketch, cmap = 'gray')
plt.axis('off')
plt.title('pencil image')
plt.show()
```
